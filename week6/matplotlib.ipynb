### 데이터 시각화

```python
import matplotlib.pyplot as plt
import numpy as np

years = np.linspace(1950, 2010, 7)
gdp = np.array([67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3])
years = list(years)
gdp = list(gdp)

plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
plt.title('GDP per captia')
plt.ylabel('dollars')
plt.savefig('gdp_per_capita.png', dpi=600)
plt.show()
```

```python
plt.plot(gdp, years, color='red', marker='o', linestyle='solid')
plt.title('GDP per capita')
plt.ylabel('year')
plt.xlabel('GDP')
plt.show()
```

```python
x = [x for x in range(-20, 20)]
y1 = [2*t for t in x]
y2 = [t**2 + 5 for t in x]
y3 = [-t**2 - 5 for t in x]

plt.plot(x, y1, 'r--', x, y2, 'g^-', x, y3, 'b*-')
plt.xlim([-30, 30])
plt.ylim([-30, 30])
plt.show()
```

```python
x = [x for x in range(20)]
y = [x**2 for x in range(20)]
z = [x**3 for x in range(20)]

plt.plot(x, x, label='linear')
plt.plot(x, y, label='quadratic')
plt.plot(x, z, label='cubic')
plt.legend()
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('My plot')
plt.show()
```

- `plt.bar()`     : 막대 차트
- `plt.scatter()` : 산포도 플롯
- `plt.pie()`     : 파이 차트
- `plt.hist()`    : 히스토그램
- `plt.boxplot()` : 상자 차트

### subplots()
- 한 화면에 여러 그래프를 동시에 시각화하는 함수

```python
fig, ax = plt.subplots(2, 2, figsize=(5, 5))

ax[0, 0].plot(range(10), 'r')
ax[0, 1].plot(range(10), 'g')
ax[1, 0].plot(range(10), 'b')
ax[1, 1].plot(range(10), 'k')
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

# 각 서브플롯에 그릴 데이터 생성
x = np.linspace(0, 10, 100)
y1 = x
y2 = x ** 2
y3 = np.sqrt(x)
y4 = np.log(x + 1)  # log(0) 방지를 위해 +1

# 각 서브플롯에 선형 그래프 그리기
ax1.plot(x, y1, 'r')  # 빨간색 선
ax2.plot(x, y2, 'g')  # 초록색 선
ax3.plot(x, y3, 'b')  # 파란색 선
ax4.plot(x, y4, 'k')  # 검은색 선

# 서브플롯 제목 설정
ax1.set_title("Linear")
ax2.set_title("Quadratic")
ax3.set_title("Square root")
ax4.set_title("Logarithmic")

# 전체 그래프의 타이틀 설정
fig.suptitle('Multiple Line Plots')

# 레이아웃 조정
plt.tight_layout()
plt.show()
```
